# 2. 加入securities.csv~

securities.csv包含了这些股票的基本信息

1. 请列举出各个sector中的加入时间最早的股票名称（10分）
2. 请列举出每一个州中加入时间最晚的股票名称（10分）


```python
df2 = pd.read_csv('securities.csv')

#导入数据
print('第一题')

#第一题请列举出各个sector中的加入时间最早的股票名称

sector = list(set(df2['GICS Sector']))


for i in sector:
    df3 = df2[df2['GICS Sector']==i][['Ticker symbol','Date first added']]
    #使用DatetimeIndex提取年月日，过滤年月日中的最小值既是最早加入的股票
    df3['year'] = pd.DatetimeIndex(df3['Date first added']).year
    df3['month'] = pd.DatetimeIndex(df3['Date first added']).month
    df3['day'] = pd.DatetimeIndex(df3['Date first added']).day
    df3['year'].min()
    df3_year = df3[df3['year']==df3['year'].min()][['Ticker symbol','month','day']]
    df3_year['month'].min()
    df3_month =df3_year[df3_year['month']==df3_year['month'].min()][['Ticker symbol','day']]
    df3_month.min()
    df3_day =df3_month[df3_month['day']==df3_month['day'].min()][['Ticker symbol']]
    print('最早加入',i,'的股票是','\n',df3_day[['Ticker symbol']])        



print('第二题', '\n')
#第二题请列举出每一个州中加入时间最晚的股票名称
state = list(set(df2['Address of Headquarters']))
for i in state:
    df4 = df2[df2['Address of Headquarters']==i][['Ticker symbol','Date first added']]
    #使用DatetimeIndex提取年月日，过滤年月日中的最小值既是最早加入的股票
    df4['year'] = pd.DatetimeIndex(df4['Date first added']).year
    df4['month'] = pd.DatetimeIndex(df4['Date first added']).month
    df4['day'] = pd.DatetimeIndex(df4['Date first added']).day
    df4['year'].max()
    df4_year = df4[df4['year']==df4['year'].max()][['Ticker symbol','month','day']]
    df4_year['month'].min()
    df4_month =df4_year[df4_year['month']==df4_year['month'].max()][['Ticker symbol','day']]
    df4_month.min()
    df4_day =df4_month[df4_month['day']==df4_month['day'].max()][['Ticker symbol']]
    print('最晚加入',i,'的股票是','\n',df4_day[['Ticker symbol']])  



    
```

```python
 #筛选出2013-2016年的数据
df6 = df5[ df5['For Year'] > 2012.0]
df7 = df6[ df6['For Year'] < 2017.0]

Sector_sum = df7.groupby('GICS Sector')['Research and Development'].sum()
print('每个sector在2013-2016年累计Research&Development的总投入:\n',Sector_sum)


#第二题没时间了就copy王同学的
df8 = df7.set_index(keys=['GICS Sector'])
col_n = ['Security','For Year','Research and Development']
df8 = pd.DataFrame(df8,columns = col_n) #刚发现的去掉多余列的方法……

sector2 = {} #key=sector, value=number of company
company = {} #key=company, value=sector
number = {} #key=company, value=number
least = {} #key=sector, value=company
com = ''#临时变量
for i in range(0,len(df8.index),1):
    
    #公司数不够3个
    if df8.index[i] not in sector2.keys(): # 新的sector
        sector2[df8.index[i]] = 1
        com = df8.iat[i,0]
        company[com] = df8.index[i]
        number[com] = float(df8.iat[i,2])
        least[df8.index[i]] = com   
    elif sector2[df8.index[i]] < 3 and df8.iat[i,0] == com and (i+1 == len(df8.index) or df8.iat[i+1,0] == com): #累加年份
        number[df8.iat[i,0]] +=  float(df8.iat[i,2])   
    elif sector2[df8.index[i]] < 3 and df8.iat[i,0] == com and (i+1 == len(df8.index) or not(df8.iat[i+1,0] == com)): #设置最菜公司
        number[df8.iat[i,0]] +=  float(df8.iat[i,2])
        if number[df8.iat[i,0]] < number[least[df8.index[i]]]:
            least[df8.index[i]] = com
    elif sector2[df8.index[i]] < 3: #同sector新公司
        sector2[df8.index[i]] += 1
        com = df8.iat[i,0]
        company[com] = df8.index[i]
        number[com] = float(df8.iat[i,2])
    
    #公司数多于3个
    elif sector2[df8.index[i]] >= 3 and not(df8.iat[i,0] in number.keys()): #新公司
        com = df8.iat[i,0]
        number[com] = float(df8.iat[i,2])
    elif sector2[df8.index[i]] >= 3 and df8.iat[i,0] == com and (i+1 == len(df8.index) or df8.iat[i+1,0] == com):#累加年份
         number[df8.iat[i,0]] +=  float(df8.iat[i,2])
    elif sector2[df8.index[i]] >= 3 and df8.iat[i,0] == com and (i+1 == len(df8.index) or not(df8.iat[i+1,0] == com)): #与最菜比较并更新最菜
        number[df8.iat[i,0]] +=  float(df8.iat[i,2])
        if number[df8.iat[i,0]] > number[least[df8.index[i]]]:
            del company[least[df8.index[i]]]
            company[com] = df8.index[i]
            tem_least = com
            for k,v in company.items(): #寻找下一个最菜
                if v == df8.index[i] and number[k] < number[tem_least]:
                    tem_least = k
            least[df8.index[i]] = tem_least #将临时最菜当做真的最菜

print('\n每个sector中，在2013-2016年累计Research&development投入最大的3家公司的名称以及投入的数值:')
for k,v in sector2.items():
    print('Sector: ',k)
    for ke,va in company.items():
        if va == k:
            print('company: ',ke,'投入的数值: ',number[ke])
    print()
```
